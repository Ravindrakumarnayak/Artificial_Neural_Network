```python

# first neural network with keras tutorial

from pandas import read_csv # for reading csv file
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense # for number of hidden layers
import tensorflow as tf
import random as rn
import os
import numpy as np
os.environ['PYTHONHASHSEED'] = '0'
np.random.seed(37)
rn.seed(1254)
tf.random.set_seed(89)


# load the dataset

filename = 'pima-indians-diabetes.data.csv'
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = read_csv(filename, names=names)
array = dataframe.values

# split into input (X) and output (y) variables

X = array[:,0:8] # independed variables
y = array[:,8] # dependent variables # checking who is diabetic and who is not

# define the keras model

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))


# compile the keras model

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# fit the keras model on the dataset

history = model.fit(X, y, epochs=150,validation_split=0.33, batch_size=10)
# fitting model into x and y parameters, epochs for number of times it will run to imporove accuracy and reducing loss in it.
#validation_split is for splitting data into  testing data size 33% and 67% training data  and batch is mini batch since it has taken value 10..

# evaluate the keras model


scores = model.evaluate(X, y) # we get accuracy around 73% in this case

print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# now ploting the model


# list all data in history

print(history.history.keys())

# summarize history for accuracy

import matplotlib.pyplot as plt
plt.plot(history.history['accuracy']) # training data plotting
plt.plot(history.history['val_accuracy']) # testing data plotting
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()

```

```python
# summarize history for loss

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
